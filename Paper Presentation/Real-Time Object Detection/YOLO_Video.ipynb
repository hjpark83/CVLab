```python
import torch

model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
import cv2

# ë™ì˜ìƒ íŒŒì¼ ì—´ê¸°
video_path = 'EU2024_final.mp4'
cap = cv2.VideoCapture(video_path)

# ë™ì˜ìƒ ì €ì¥ ì„¤ì •
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_video.mp4', fourcc, 30.0, (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # í”„ë ˆì„ì—ì„œ ê°ì²´ ê°ì§€
    results = model(frame)

    # ê²°ê³¼ë¥¼ ì´ë¯¸ì§€ì— ê·¸ë¦¬ê¸°
    frame = results.render()[0]

    # ê²°ê³¼ í”„ë ˆì„ì„ ë™ì˜ìƒìœ¼ë¡œ ì €ì¥
    out.write(frame)

    # í™”ë©´ì— ê²°ê³¼ í”„ë ˆì„ í‘œì‹œ (ì„ íƒ ì‚¬í•­)
    # cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# ìì› í•´ì œ
cap.release()
out.release()
cv2.destroyAllWindows()
```
